```python
######EJERCICIO 1#################################
import numpy as np
import matplotlib.pyplot as plt

# Método de Euler
def euler_method(f, t0, y0, h, t_end):
    n_steps = int((t_end - t0) / h) + 1
    t_values = np.linspace(t0, t_end, n_steps)
    y_values = np.zeros(n_steps)
    y_values[0] = y0

    for i in range(1, n_steps):
        y_values[i] = y_values[i-1] + h * f(t_values[i-1], y_values[i-1])
    
    return t_values, y_values

# Literal a: y' = t * e^(3t) - 2y
def f1a(t, y):
    return t * np.exp(3 * t) - 2 * y

# Parámetros del problema
t_values, y_values = euler_method(f1a, t0=0, y0=0, h=0.5, t_end=1)

# Graficar resultados
plt.plot(t_values, y_values, 'o-', label="Aproximación (Euler)")
plt.title("Ejercicio 1a: Método de Euler")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Método de Euler
def euler_method(f, t0, y0, h, t_end):
    n_steps = int((t_end - t0) / h) + 1
    t_values = np.linspace(t0, t_end, n_steps)
    y_values = np.zeros(n_steps)
    y_values[0] = y0

    for i in range(1, n_steps):
        y_values[i] = y_values[i-1] + h * f(t_values[i-1], y_values[i-1])

    return t_values, y_values

# Solución exacta para y' = t + y
def exact_solution(t):
    return np.exp(t) + t - 1

# Definir la función para y' = t + y
def f1a(t, y):
    return t + y

# Parámetros del problema
t_values, y_approx = euler_method(f1a, t0=0, y0=1, h=0.1, t_end=1)

# Solución exacta
y_exact = exact_solution(t_values)

# Calcular el error absoluto
error_abs = np.abs(y_exact - y_approx)

# Graficar la comparación y el error
plt.plot(t_values, y_exact, label="Solución exacta")
plt.plot(t_values, y_approx, 'o-', label="Aproximación (Euler)")
plt.title("Comparación: Solución exacta vs Método de Euler")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.grid()
plt.legend()
plt.show()

# Graficar el error absoluto
plt.plot(t_values, error_abs, 'o-', label="Error absoluto")
plt.title("Error absoluto: Método de Euler vs Solución exacta")
plt.xlabel("t")
plt.ylabel("Error")
plt.grid()
plt.legend()
plt.show()
```

```python
######EJERCICIO 6#################################
# Método de Taylor de orden 2
def taylor_order2(f, dfdt, dfdy, t0, y0, h, t_end):
    n_steps = int((t_end - t0) / h) + 1
    t_values = np.linspace(t0, t_end, n_steps)
    y_values = np.zeros(n_steps)
    y_values[0] = y0

    for i in range(1, n_steps):
        t = t_values[i-1]
        y = y_values[i-1]
        y_values[i] = y + h * f(t, y) + 0.5 * h**2 * (dfdt(t, y) + dfdy(t, y) * f(t, y))
    
    return t_values, y_values
```

```python

# Literal a: y' = t * e^(3t) - 2y
def f6a(t, y):
    return t * np.exp(3 * t) - 2 * y

# Derivadas parciales de f con respecto a t e y
def df6a_dt(t, y):
    return np.exp(3 * t) * (1 + 3 * t)

def df6a_dy(t, y):
    return -2

# Parámetros del problema
t_values, y_values = taylor_order2(f6a, df6a_dt, df6a_dy, t0=0, y0=0, h=0.5, t_end=1)

# Graficar resultados
plt.plot(t_values, y_values, 'o-', label="Aproximación (Taylor orden 2)")
plt.title("Ejercicio 6a: Método de Taylor de orden 2")
plt.xlabel("t")
plt.ylabel("y(t)")
plt.grid()
plt.legend()
plt.show()
```

```python
######EJERCICIO 7#################################
# Método de Taylor de orden 4
def taylor_order4(f, dfdt, dfdy, d2fdt2, d2fdy2, d2fdtdy, t0, y0, h, t_end):
    n_steps = int((t_end - t0) / h) + 1
    t_values = np.linspace(t0, t_end, n_steps)
    y_values = np.zeros(n_steps)
    y_values[0] = y0

    for i in range(1, n_steps):
        t = t_values[i-1]
        y = y_values[i-1]
        y_values[i] = y + h * f(t, y) + (h**2 / 2) * (dfdt(t, y) + dfdy(t, y) * f(t, y)) + \
                        (h**3 / 6) * (d2fdt2(t, y) + d2fdy2(t, y) * f(t, y) + d2fdtdy(t, y)) + \
                        (h**4 / 24) * (d2fdt2(t, y) * f(t, y))
    
    return t_values, y_values
```
